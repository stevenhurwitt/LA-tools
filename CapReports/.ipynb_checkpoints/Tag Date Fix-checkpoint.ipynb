## Cap Tag Date Fix

```python
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize
import datetime as dt
import json
import cx_Oracle
import os
import CapTags as ct
```

```python
PRs = pd.read_csv('NEPOOL_queue_offer.csv', header = None, dtype = 'str')
PRs.columns = ['PR', 'rev']

pr_list = ["_".join([a, b]) for a, b in zip(PRs.PR, PRs.rev)]
pr_list
```

```python
master = pd.DataFrame()

for pr in pr_list:
    
    try:
        report = ct.get_report(pr)
        new = ct.fix_my_PR(report)
    
        master = pd.concat([master, new], axis = 0)
        
    except:
        print('no info for', pr)
        
master = master.reset_index(drop = True)
```

```python
for index, row in master.iterrows():
    
    update = "UPDATE pwrline.lscmcontract"
    
    set_state = "".join(["SET starttime = ", str(row['right start']), ' and stoptime = ', str(row['right stop'])])
    
    where = "".join(["WHERE accountid = ", str(row.AccountID), ", and contractid = ", str(row.ContractID), ", and revision = ", str(row.Revision)])
    
    tag_query = " ".join([update, set_state, where])
    
    print(tag_query)
    
```
